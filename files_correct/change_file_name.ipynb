```python
# 导入必要的库
import os 
import glob
import re
```

```python
def modify_filename(old_name, new_name): # 修改文件名称
    os.rename(old_name, new_name)
```

```python
def remove_chinese(string):
    pattern = r'[\u4e00-\u9fa5]+' # 匹配所有中文字符
    result = re.sub(pattern, '', string) # 将匹配到的中文字符替换为空字符串
    return result
```

```python
path = r'D:/Transmorph_Frame/Data/'
file_names_old = glob.glob(path + 'raw_tumors/' + '*') # 读取完整数据
print(file_names_old)
file_names_new = file_names_old.copy()
```

```python
for i in range(len(file_names_old)):
    modify_filename(file_names_old[i], remove_chinese(file_names_old[i]))
```
